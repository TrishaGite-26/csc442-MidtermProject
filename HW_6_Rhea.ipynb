```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as mpl
```

```python
df = pd.read_csv('heart_disease.csv')

print(df.head())
```

```python
df.shape
```

```python
avg_bmi_by_disease = df.groupby('heart_disease')['bmi'].mean()

# Display result
print("Average BMI by heart disease status:")
print(avg_bmi_by_disease)
```

```python
mpl.figure(figsize=(10, 6))
sns.boxplot(x='heart_disease', y='bmi', data=df)
mpl.title('BMI Distribution by Heart Disease Status')
mpl.xlabel('Heart Disease (0 = No, 1 = Yes)')
mpl.ylabel('BMI')
mpl.show()
```

```python
heart_disease_only = df[df["heart_disease"] == 1]

# Set figure size
plt.figure(figsize=(12, 6))

# Plot frequency chart of BMI for heart disease cases
sns.histplot(data=heart_disease_only, x="bmi", bins=30, color="red")

# Labels and title
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.title("Frequency of Heart Disease Cases by BMI")

# Show plot
plt.show()
```

```python
correlation = df['bmi'].corr(df['heart_disease'])
print(f'Correlation between BMI and heart disease: {correlation}')
```
