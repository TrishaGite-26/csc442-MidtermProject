# States with Highest Heart Disease Risk

*Data Querying and Exploratory data analysis and visualization*

This code analyzes heart disease prevalence across states by first calculating the average heart disease rate for each state. It then sorts these averages in descending order to identify states with the highest prevalence. The results are displayed in a table showing the top states and their rates. Additionally, a bar chart visualizes the top 10 states, and a box plot illustrates the distribution of heart disease rates across all states. The performance of the query is also measured, displaying execution time and total rows returned.

What it does: Imports necessary libraries and loads the dataset from a CSV file.

Relation to query: I began the analysis by loading data relevant to heart disease prevalence across different states, essential for identifying areas at higher risk.

```python
import pandas as pd
import time
```

What it does: Groups the data by state and calculates the average heart disease rate for each, then sorts the results in descending order.

Relation to query: This step is critical to answering the query by identifying states with the highest average heart disease prevalence, which directly answers "Which states are more prevalent to heart disease risk?"

```python
# Load the dataset
df = pd.read_csv("Heart_Disease_and_Hospitals.csv")
```

```python
# Start timing the query execution
start_time = time.time()

# Group by state and calculate the average heart disease prevalence
state_heart_disease = df.groupby("state")["heart_disease"].mean().reset_index()

# Number of rows in the result
# num_rows = state_heart_disease.shape[0]
num_rows_original = df.shape[0]

# Number of columns in the result
#num_columns_original = df.shape[1]
num_columns_original = df.shape[1]
```

```python
# Sort in descending order to find states with the highest prevalence
state_heart_disease = state_heart_disease.sort_values(by="heart_disease", ascending=False)
```

What it does: Measures the execution time of the query, displays the first two results, and shows how many rows were returned.

Relation to query: This provides performance metrics (time) and quick insights into the highest heart disease prevalence, confirming that the query was successful and efficient.

```python
# End timing
end_time = time.time()
```

The output below show the states "Wyoming" and "Oregon" along with their average heart disease prevalence rates. Wyoming has a rate of 0.569892, while Oregon has a rate of 0.548246. These values represent the average percentage of people in each state with heart disease, with the higher number indicating a slightly higher prevalence in Wyoming compared to Oregon.

```python
# Display the first two rows of results
print(state_heart_disease.head(2))

#Display total rows
print(f"Total rows returned: {num_rows_original }")

#Display total columns
print(f"Total columns returned: {num_columns_original}")
```

```python
# Print execution time and total rows returned
print(f"Total rows returned: {state_heart_disease.shape[0]}")
print(f"Query execution time: {end_time - start_time:.4f} seconds")
```

What it does: Creates a bar chart visualizing the top 10 states with the highest heart disease prevalence.

Relation to query: This plot visually supports the query by highlighting the states most affected by heart disease, allowing for easy comparison.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot bar chart for top 10 states with highest heart disease prevalence
plt.figure(figsize=(12,6))
sns.barplot(x=state_heart_disease["state"][:10], y=state_heart_disease["heart_disease"][:10], palette="Reds")
plt.xticks(rotation=45)
plt.xlabel("State")
plt.ylabel("Average Heart Disease Rate (%)")
plt.title("Top 10 States with Highest Heart Disease Prevalence")
plt.show()
```

What it does: Generates a box plot to visualize the distribution of heart disease prevalence rates across all states.

Relation to query: The box plot provides a broader view of the variability in heart disease rates, showing how states differ in their prevalence and identifying outliers. This gives deeper context to which states face higher or lower risks overall.
